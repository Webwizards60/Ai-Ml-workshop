```python
#sample matplotlib diagram
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 20)
y = x**2
plt.plot(x, y)
```

```python
#time and distance graph
plt.title("Cubic relationship between x and y")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y)
```

```python
#Formatting line type and color
x = np.arange(0, 20)
y = x**3
plt.title("Cubical relation between x and y")
plt.xlabel("Distance")
plt.ylabel("Time")
plt.plot(x, y, "r") # Formatting the line colors
plt.plot(x, y, "|") # Formatting the line type
plt.savefig("graph.pdf", format="pdf")
```

```python
#adding multiple annotations
z = x^3
t = x^4

plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y)
#Annotate
plt.annotate('second Entry', xy=[2, 1])
plt.annotate('Third Entry', xy = [4, 6])
#Adding legends
plt.plot(x, z)
plt.plot(x, t)
plt.legend(['Race1', 'Race2', 'Race3'])
```

SUPERIMPOSING 2 GRAPHS

```python
# SUPERIMPOSING 2 GRAPHS
%matplotlib inline
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(1.0, 3.0, 0.5)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.plot(t1, f(t1), "g", t2, f(t2), "r")
```

PLOTTING PIE GRAPH

```python
#plotting pie graph
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
